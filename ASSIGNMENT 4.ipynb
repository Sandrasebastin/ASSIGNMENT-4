```python
import pandas as pd
data=pd.read_csv(r"C:\Users\SANDRA\downloads\titanic_dataset.CSV")
data
```

```python
data['Age']=data['Age'].fillna(data['Age'].mean())
data
```

```python
data=data.fillna(1)
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
columns_to_check=['Age','Fare','SibSp']
```

```python
for col in columns_to_check:
    sns.boxplot(x=data[col])
    plt.title(f'Box Plot for {col}')
    plt.show()
    
```

```python
outliers={}
for col, outlier_df in outliers.items():
    print(f'Potential outliers in {col}:')
    print(outlier_df)
```

```python
####using IQR method,remove the outliers from the data
```

```python
for col in columns_to_check:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    data[col] = data[col].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
```

```python
for col in columns_to_check:
    sns.boxplot(x=data[col])
    plt.title(f'Box Plot for {col}')
    plt.show()
```

```python
#### no outliers
```
